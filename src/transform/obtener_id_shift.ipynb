```python
from datetime import datetime

from backports import zoneinfo

# Ejemplo con los valores que recibes
started_at_utc = "2025-07-04T18:03:35+00:00"
ended_at_utc = ""

# Convertir a datetime y luego a zona Chile
dt_inicio_utc = datetime.fromisoformat(started_at_utc)
dt_inicio_chile = dt_inicio_utc.astimezone(zoneinfo.ZoneInfo("America/Santiago"))
print("Hora de inicio en Chile:", dt_inicio_chile.strftime("%d/%m/%Y %H:%M"))

if ended_at_utc == "":
    print("El shift aún no ha terminado.")
else:
    # Convertir la hora de cierre a Chile
    dt_cierre_utc = datetime.fromisoformat(ended_at_utc)
    dt_cierre_chile = dt_cierre_utc.astimezone(zoneinfo.ZoneInfo("America/Santiago"))
    print("Hora de cierre en Chile:", dt_cierre_chile.strftime("%d/%m/%Y %H:%M"))
```
